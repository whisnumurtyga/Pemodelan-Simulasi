```python
from modsim import * 
import random
```

```python
def food_stall_simulation(num_orders):
    orders = []
    total_time = 0

    for order in range(1, num_orders + 1):
        num_food = random.randint(1, 5)  # Jumlah makanan yang dipesan
        difficulty = random.choice(['A', 'B', 'C'])  # Tingkat kesulitan makanan

        if difficulty == 'A':
            process_time = random.randint(5, 10)  # Waktu pemrosesan makanan
            color = 'red'  # Warna untuk tingkat kesulitan A
        elif difficulty == 'B':
            process_time = random.randint(10, 20)
            color = 'blue'  # Warna untuk tingkat kesulitan B
        else:
            process_time = random.randint(20, 30)
            color = 'green'  # Warna untuk tingkat kesulitan C

        completion_time = total_time + process_time
        orders.append((order, num_food, difficulty, process_time, completion_time, color))
        total_time += process_time

    return orders
```

```python
def simmulation(num_orders) :
    # Menjalankan simulasi
    orders = food_stall_simulation(num_orders)

    # Menyiapkan data untuk plot
    order_numbers = [order[0] for order in orders]
    num_foods = [order[1] for order in orders]
    completion_times = [order[4] for order in orders]
    colors = [order[5] for order in orders]

    # Plot hasil simulasi
    plt.figure(figsize=(10, 6))
    plt.bar(order_numbers, num_foods, label='Num Food')
    plt.bar(order_numbers, completion_times, color=colors, alpha=0.7, label='Completion Time')
    plt.xlabel('Order Number')
    plt.ylabel('Quantity / Time')
    plt.title('Food Stall Simulation')
    plt.legend()
    plt.show()
```

```python
simmulation(20)
```
