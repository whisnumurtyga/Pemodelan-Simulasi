```python
import matplotlib.pyplot as plt

def simulate_portfolio_growth(initial_investment, growth_rate, loss_rate, num_years):
    portfolio = [initial_investment]
    for _ in range(num_years):
        if portfolio[-1] > 0:
            new_investment = portfolio[-1] * (1 + growth_rate)
        else:
            new_investment = portfolio[-1] * (1 + loss_rate)
        portfolio.append(new_investment)
    return portfolio

def plot_portfolio_growth(portfolio):
    years = range(len(portfolio))

    plt.plot(years, portfolio)
    plt.xlabel('Years')
    plt.ylabel('Portfolio Value')
    plt.title('Portfolio Growth')

    plt.show()

initial_investment = 1000
growth_rate = 0.05
loss_rate = -0.02
num_years = 10

portfolio = simulate_portfolio_growth(initial_investment, growth_rate, loss_rate, num_years)
plot_portfolio_growth(portfolio)
```
