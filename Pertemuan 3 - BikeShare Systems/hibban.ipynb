```python
sepeda = State(ketintang=10, gayungan=2, benowo=24)

def bike_from_ketintang_to_gayungan():
    print('Moving a bike from Ketintang to Gayungan')
    sepeda.ketintang -= 1
    sepeda.gayungan += 1

def bike_from_ketintang_to_benowo():
    print('Moving a bike from Ketintang to Benowo')
    sepeda.ketintang -= 1
    sepeda.benowo += 1

def bike_from_gayungan_to_ketintang():
    print('Moving a bike from Gayungan to Ketintang')
    sepeda.gayungan -= 1
    sepeda.ketintang += 1

def bike_from_gayungan_to_benowo():
    print('Moving a bike from Gayungan to Benowo')
    sepeda.gayungan -= 1
    sepeda.benowo += 1

def bike_from_benowo_to_ketintang():
    print('Moving a bike from Benowo to Ketintang')
    sepeda.benowo -= 1
    sepeda.ketintang += 1

def bike_from_benowo_to_gayungan():
    print('Moving a bike from Benowo to Gayungan')
    sepeda.benowo -= 1
    sepeda.gayungan += 1

def langkah():
    if flip(0.5):
        if flip(0.5):
            bike_from_ketintang_to_gayungan()
        else:
            bike_from_ketintang_to_benowo()

    if flip(0.33):
        if flip(0.5):
            bike_from_gayungan_to_ketintang()
        else:
            bike_from_gayungan_to_benowo()

    if flip(0.2):
        if flip(0.5):
            bike_from_benowo_to_ketintang()
        else:
            bike_from_benowo_to_gayungan()
```

```python
results = TimeSeries()
for i in range(10):
    print(i)
    langkah()
    results[i+1] = sepeda.ketintang
results.plot()

decorate(title='ketintang bikeshare',
         xlabel='Time step (min)', 
         ylabel='Number of bikes')
```
