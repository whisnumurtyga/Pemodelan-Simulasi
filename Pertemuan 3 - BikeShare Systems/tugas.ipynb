```python
from modsim import *
```

```python
bikeshare = State(ketintang=11, gayungan=20, benowo=5)

# CODE 
# ketintang => KTT 
# gayungan => GYG
# benowo => BNW
```

```python
def ktt_to_gyg() :
    if bikeshare.ketintang < 1 :
        print('sepeda kampus ketintang kosong, harap tunggu...')
    else :
        bikeshare.ketintang -= 1
        bikeshare.gayungan += 1

def ktt_to_bnw() :
    if bikeshare.ketintang < 1 :
        print('sepeda kampus ketintang kosong, harap tunggu...')
    else :
        bikeshare.ketintang -= 1
        bikeshare.benowo += 1

def gyg_to_ktt() :
    if bikeshare.gayungan < 1 :
        print('sepeda kampus gayungan kosong, harap tunggu...')
    else :
        bikeshare.gayungan -= 1
        bikeshare.ketintang += 1

def gyg_to_bnw() :
    if bikeshare.gayungan < 1 :
        print('sepeda kampus gayungan kosong, harap tunggu...')
    else :
        bikeshare.gayungan -= 1
        bikeshare.benowo += 1

def bnw_to_ktt() :
    if bikeshare.benowo < 1 :
        print('sepeda kampus benowo kosong, harap tunggu...')
    else :
        bikeshare.benowo -= 1
        bikeshare.ketintang += 1
    
def bnw_to_gyg() :
    if bikeshare.benowo < 1 :
        print('sepeda kampus benowo kosong, harap tunggu...')
    else :
        bikeshare.benowo -= 1
        bikeshare.gayungan += 1

def step(ktt, gyg, bnw):
    if flip(ktt) :
        if flip(ktt) :
            ktt_to_gyg()
        else:
            ktt_to_bnw()
    if flip(gyg) :
        if flip(gyg) :
            gyg_to_ktt()
        else :
            gyg_to_bnw()
    if flip(bnw) :
        if flip(bnw) :
            bnw_to_gyg()
        else :
            bnw_to_ktt()

def printAll() :
    print(f'ketintang : {bikeshare.ketintang}')
    print(f'gayungan : {bikeshare.gayungan}')
    print(f'benowo : {bikeshare.benowo}')

def simulations(n, p1, p2, p3) :
    ketintang = TimeSeries()
    gayungan = TimeSeries()
    benowo = TimeSeries()

    for i in range(n):
        print(i)
        step(p1, p2, p3)
        ketintang[i+1] = bikeshare.ketintang
        gayungan[i+1] = bikeshare.gayungan
        benowo[i+1] = bikeshare.benowo

    ketintang.plot()
    gayungan.plot()
    benowo.plot()

    decorate(title='Keintang - Gayungan - Benowo',
            xlabel='Time step (min)', 
            ylabel='Number of bikes')
    
    printAll()
```

```python
simulations(480, 0.5, 0.33, 0.2)
```
